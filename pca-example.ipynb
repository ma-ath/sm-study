```python
# simply plot images
fig, axes = plt.subplots(2, 5, figsize=(1.5*5,2*2))
for i in range(10):
    ax = axes[i//5, i%5]
    ax.imshow(mnist[i][0], cmap='gray')
    ax.set_title('Label: {}'.format(mnist[i][1]))
    ax.axis('off')

plt.tight_layout()
plt.show()
```
